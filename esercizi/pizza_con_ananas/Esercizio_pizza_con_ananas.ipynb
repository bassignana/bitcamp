```python
#
# Esercizio: ai clienti di questa pizzeria, piace la pizza con l'ananas?
# Utilizzando Pandas, rispondete alla domanda sopra. La domanda è volutamente aperta,
# quindi come prima cosa vi consiglio di cercare di capire quali variabili nel dataset
# potrebbero essere utili.
#
from pathlib import Path
import pandas as pd

# Caricare il file excel caricato su Colab.
# Sulle vostre macchine dovete comunque fornire un path.
BASE_DIR = Path(__file__).resolve().parent if '__file__' in globals() else Path.cwd()
file_path = BASE_DIR / 'Data Model - Pizza Sales.xlsx'

df = pd.read_excel(file_path)
df.head()
```

```python
# Come prima operazione.
df.info()
```

```python
print(" La top ten delle pizze più vendute è: ".center(55,'='))
top_10_pizzas = df.groupby('pizza_name')['quantity'].sum().sort_values(ascending=False).head(10)
# max_len è una variabile che contiene la lunghezza massima del nome della pizza che mi serve per formattare sotto la stampa
max_len = max(len(name) for name in top_10_pizzas.index)
for i,( pizza, qty) in enumerate(top_10_pizzas.items(),start=1):
    print(f"{i:2d}. {pizza:<{max_len+1}} | {qty} pizze vendute")
```

```python
print(" Quante Pizze totali sono state vendute? ".center(55,'='))
pizze_vendute = df['quantity'].sum()
print(f"Sono state vendute in totale {pizze_vendute:,} pizze")
```

```python
print(" Quante Pizze con l'Ananas sono state vendute? ".center(55,'='))

hawaiian_quantity = df[df['pizza_name'] == 'The Hawaiian Pizza']['quantity'].sum()
print(f"Sono state vendute in totale {hawaiian_quantity:,} pizze con l'ananas")
```

```python
pizze_totali_vendute = df['quantity'].sum()
totale_incasso = df['total_price'].sum()
tot_incasso_hawaiian = df[df['pizza_name'] == 'The Hawaiian Pizza']['total_price'].sum()
percentuale_incasso_hawaiian = (tot_incasso_hawaiian / totale_incasso) * 100

print(" Dettagli della pizza Hawaiian: ".center(55,'='))
print(f"Pizza Hawaiian quantità: {hawaiian_quantity} (pari al {percentuale_incasso_hawaiian:.2f}% di {pizze_totali_vendute:,} pizze vendute)")
print(f"Pizza Hawaiian incasso: ${tot_incasso_hawaiian:,.2f} (pari al {percentuale_incasso_hawaiian:.2f}% di ${totale_incasso:,.2f})")
```
